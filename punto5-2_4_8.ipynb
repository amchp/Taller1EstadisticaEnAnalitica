```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college = pd.read_csv('datos/College.csv')
college = college.rename ({'Unnamed: 0': 'College'}, axis =1)
college = college.set_index('College')
print(college)
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
college.boxplot('Outstate', 'Private')
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,0.5,1], labels =['No', 'Yes'])

college['Elite'].value_counts('Elite')
```

```python
college.plot.hist(subplots=True)
```
